# **Sürekli Zamanlı Sinyallerin Gösterimi**

Fonksiyonun bağımsız değişkeni **t** ile gösterilen zamansa ve **g(t)** fonksiyonu her **t** değeri için tanımlıysa, bu fonksiyon **sürekli-zamanlı fonksiyon** olarak adlandırılır.

image.png

## Sinüs Sinyali

```python
def sin(A, f):

  t = np.linspace(-3,3,200)
  sinyal = A* np.sin(2*np.pi*f*t)

  plt.plot(t, sinyal, color='#FF5733')

  return
```

```python
sin(4, 100)
```

## Kosinüs Sinyali

```python
def cos(A, f):

  t = np.linspace(-3,3,200)
  sinyal = A* np.cos(2*np.pi*f*t)

  plt.plot(t, sinyal, color='#7D3C98')

  return
```

```python
cos(-2, 100)
```

## Üstel Sinyal

```python
def exp(A, f):

  t = np.linspace(1,3,100)
  sinyal = A* np.exp(-f*t)

  plt.plot(t, sinyal, color='#1A5276')

  return
```

```python
exp(2, 5)
```

## Karmaşık Üstel

```python
t = np.linspace(0,np.pi,200)
f = 5
sigma = 2
omega = 2*np.pi*f
s = sigma + 1j*omega
y = np.exp(s*t)
plt.plot(t,y, color='#EC7063')
```

# **Sürekli Sinyallerin Tekillik Fonksiyonları**

## Signum Fonksiyonu

```python
t = np.linspace(-1, 1, 500)
y = np.sign(t)

plt.plot(t,y, color='#9B59B6')
```

## Birim Basamak Fonksiyonu

```python
t = np.linspace(-1, 1, 500)
y = np.heaviside(t,1)
plt.plot(t,y, color='#D98880')
```

## Birim Rampa Fonksiyonu

```python
def rampa(t): 
  return np.maximum(0,t)

t = np.linspace(-1, 1, 200)
y = rampa(t)

plt.plot(t,y, color='#196F3D')
```

## Birim Dürtü

```python
t = np.linspace(-1,1,100)
y = ss.unit_impulse(len(t), idx='mid')
plt.stem(t,y)
```

## Dikdörtgen Darbe Fonksiyonu

```python
def darbe(t,tau): 
  return 1*(np.abs(t)<=tau/2)

t = np.linspace(-1, 1, 500)
y = darbe(t, 1)
plt.plot(t,y, color='#E74C3C')
```

## Üçgen Darbe Fonksiyonu

```python
def triang(t,tau): 
  return (1 - np.abs(t)/tau)*(np.abs(t)<float(tau))

t = np.linspace(-1, 1, 500)
y = triang(t, .5)
plt.plot(t,y)
```


# **Kaydırma ve Ölçekleme**

g(t) = t**2 fonksiyonu için :


```python
def signal(t): 
  return t**2

t = np.linspace(-8, 8, 100)
```

### Zamanda Kaydırma

image.png

```python
x1 = signal(t)
x2 = signal(t+2)
x3 = signal(t-3)

plt.plot(t, x1, label='t')
plt.plot(t, x2, label='t+2')
plt.plot(t, x3, label='t-3')

plt.legend()
plt.show()
```

```python
y1 = signal(t)
y2 = signal(t)+7
y3 = signal(t)-5

plt.plot(t, y1, label='t')
plt.plot(t, y2, label='t+2')
plt.plot(t, y3, label='t-3')

plt.legend()
plt.show()
```

### Genlik Ölçekleme

image.png

```python
z1 = signal(t)
z2 = 5*signal(t)
z3 = signal(t)/3
z4 = -2*signal(t)

plt.plot(t, z1, label='t')
plt.plot(t, z2, label='5t')
plt.plot(t, z3, label='t/3')
plt.plot(t, z4, label='-2t')
plt.legend()
plt.show()
```

```python
z1 = signal(t)
z2 = signal(5*t)
z3 = signal(t/3)
z4 = signal(-2*t)

plt.plot(t, z1, label='t')
plt.plot(t, z2, label='5t')
plt.plot(t, z3, label='t/3')
plt.plot(t, z4, label='-2t')
plt.legend()
plt.show()
```
